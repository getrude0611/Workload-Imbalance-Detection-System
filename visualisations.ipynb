```python
import pandas as pd

df = pd.read_csv('cleaned_helpdesk_tickets.csv')
```

# Avg resolution time by team

```python
import matplotlib.pyplot as plt

avg_time = df.groupby('assigned_team')['resolution_time_hrs'].mean()

plt.figure(figsize=(8,5))
avg_time.plot(kind='bar')
plt.title('Average Resolution Time by Team')
plt.xlabel('Assigned Team')
plt.ylabel('Average Resolution Time (Hours)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('avg_resolution_time.png')
plt.show()
```

# Escalation rate by team

```python
esc_rate = df.groupby('assigned_team')['escalated'].mean()

plt.figure(figsize=(8,5))
esc_rate.plot(kind='bar')
plt.title('Escalation Rate by Team')
plt.xlabel('Assigned Team')
plt.ylabel('Escalation Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('escation_rate_by_team.png')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x='assigned_team', y='resolution_time_hrs', data=df)
plt.title('Resolution Time Distribution by Team')
plt.xticks(rotation=45)
plt.savefig('resolution_time_distribution')
plt.show()
```
